```python
import json
import pandas as pd
```

```python
def classificar_theta(theta, limiar):
    if theta <= -limiar:
        return "baixa"
    elif -limiar < theta <= limiar:
        return "media"
    else:
        return "alta"

limiares = [0.5, 1, 1.5]

resultados = {}
```

```python
path = r"D:\ODG\outputs n8n\Prompt 2 (Green)\Prompt 2 em Lotes + ID.json"

with open(path, "r", encoding="utf-8") as f:
    data = json.load(f)
```

### Prompt Green

```python
df = pd.json_normalize(data)
df
```

```python
df_students = df.explode("output.students")

df_students["theta"] = df_students["output.students"].apply(lambda x: x["theta"])
df_students["score"] = df_students["output.students"].apply(lambda x: x["score"])
df_students
```

```python
for limiar in limiares:
    
    df_temp = df_students.copy()
    
    df_temp["grupo"] = df_temp["theta"].apply(lambda x: classificar_theta(x, limiar))
    
    resumo = (
        df_temp
        .groupby(["output.id_questao", "grupo"])["score"]
        .agg(["count", "mean"])
        .rename(columns={"mean": "taxa_acerto"})
        .reset_index()
    )
    
    resultados[limiar] = resumo
```

```python
for limiar, tabela in resultados.items():
    print(f"\n===== LIMIAR: {limiar} =====")
    print(tabela)
```

### 3 questões com os limiares

```python
path = r"D:\ODG\outputs n8n\Prompt 2 (Green)\Prompt 2 - Questões 9 35 38.json"

with open(path, "r", encoding="utf-8") as f:
    data = json.load(f)
```

```python
questoes = pd.json_normalize(data)
questoes
```

```python
questoes_3 = questoes.explode("output.students")

questoes_3["theta"] = questoes_3["output.students"].apply(lambda x: x["theta"])
questoes_3["score"] = questoes_3["output.students"].apply(lambda x: x["score"])

#df_students = df_students.drop(columns=["students"])
questoes_3
```

```python
def classificar_theta(theta, limiar):
    if theta <= -limiar:
        return "baixa"
    elif -limiar < theta <= limiar:
        return "media"
    else:
        return "alta"


limiares = [0.5, 1, 1.5]

ordem_grupos = ["baixa", "media", "alta"]
ordem_questoes = ["fácil", "média", "difícil"]

# 🔥 MAPEAMENTO DAS QUESTÕES (AJUSTE CONFORME NECESSÁRIO)
mapa_questoes = {
    35: "fácil",
    38: "média",
    9: "difícil"
}

resultados = {}

for limiar in limiares:

    df_temp = questoes_3.copy()
    df_temp["grupo"] = df_temp["theta"].apply(lambda x: classificar_theta(x, limiar))

    df_temp["grupo"] = pd.Categorical(
        df_temp["grupo"],
        categories=ordem_grupos,
        ordered=True
    )

    df_temp["nivel_questao"] = df_temp["output.id_questao"].map(mapa_questoes)

    df_temp["nivel_questao"] = pd.Categorical(
        df_temp["nivel_questao"],
        categories=ordem_questoes,
        ordered=True
    )

    resumo = (
        df_temp
        .groupby(["nivel_questao", "output.id_questao", "grupo"], observed=True)["score"]
        .agg(["count", "mean"])
        .rename(columns={"mean": "taxa_acerto"})
        .reset_index()
        .sort_values(["nivel_questao", "output.id_questao", "grupo"])
    )

    resultados[limiar] = resumo
```

```python
for limiar, tabela in resultados.items():

    print(f"\n ===== PIVOT LIMIAR {limiar} =====")

    pivot = tabela.pivot(
        index="grupo",
        columns=["nivel_questao", "output.id_questao"],
        values="taxa_acerto"
    )

    print(pivot)
    
#esse foi legalzinho
```
