```python
import numpy as np
import json
from scipy.stats import truncnorm
from scipy.stats import norm

import matplotlib.pyplot as plt
```

```python
numero_alunos = 1000
media = 0
desvio_padrao = 1
limite_inferior = -3
limite_superior = 3

np.random.seed(57)  # garante reprodutibilidade
a = (limite_inferior - media) / desvio_padrao
b = (limite_superior - media) / desvio_padrao

dados_truncados = truncnorm.rvs(a, b, loc=media, scale=desvio_padrao, size=numero_alunos)

students = {
    "estudantes": [
        {
            "id_estudante": str(i + 1),
            "habilidade": round(float(dados_truncados[i]), 4)
        }
        for i in range(numero_alunos)
    ]
}

print(json.dumps(students, indent=2))
```

```python
with open("students_1000.json", "w", encoding="utf-8") as f:
    json.dump(students, f, indent=2, ensure_ascii=False)
```

```python
plt.figure(figsize=(8,5))

# histograma
#plt.hist(dados_truncados, bins='fd',density=True, alpha=0.6, label="Dados simulados")
plt.hist(dados_truncados,density=True, alpha=0.6, label="Dados simulados")


# curva normal teórica
x = np.linspace(-3,3,200)
plt.plot(x, norm.pdf(x, 0, 1), linewidth=2, label="Normal N(0,1)")

plt.title("Distribuição das Habilidades dos 1000 Estudantes")
plt.xlabel("Habilidade (θ)")
#plt.ylabel("Densidade")

plt.legend()
plt.grid(True)

plt.show()
```

```python
students_amostra_sorted = sorted(
    students_amostra,
    key=lambda x: int(x["id_estudante"])
)

students_amostra_sorted
```

```python
# pegar apenas as habilidades da amostra de 100
habilidades_100 = [s["habilidade"] for s in students_amostra_sorted]

plt.figure(figsize=(8,5))

# histograma da amostra
plt.hist(habilidades_100, density=True, alpha=0.6, label="Amostra de 100 estudantes")

# curva normal teórica
x = np.linspace(-3,3,200)
plt.plot(x, norm.pdf(x, 0, 1), linewidth=2, label="Normal N(0,1)")

plt.title("Distribuição das Habilidades dos 100 Estudantes (Amostra Diversa)")
plt.xlabel("Habilidade (θ)")

plt.legend()
plt.grid(True)

plt.show()
```
